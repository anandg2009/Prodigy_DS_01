Data Visulaization

```python
sns.lineplot(x='median_age',y='land_area',data=data)
plt.show()
```

```python
sns.scatterplot(x='median_age',y='land_area',data=data)
plt.title('Scatter plot')
plt.show
```

```python
# barplot
## showing which country has who much the land_area
sns.barplot(x='country',y='land_area',data=data)
plt.show()
```

```python
sns.histplot(x='region',data=data)
plt.show()
```

```python
sns.histplot(x='region',y='median_age',data=data)
plt.show()
```

```python
sns.pairplot(data=data)
plt.show()
```
